```python
import geopy.distance
import pandas as pd
```

```python
def combine_lat_long(df_obj):
    return (df_obj.lat, df_obj.long)

def get_distance(df_obj):
    try:
        return geopy.distance.distance(df_obj.latlong, df_obj.next_lat_long).m
    except ValueError:
        return 0
```

```python
df = pd.read_csv('cordinates.csv')
df.head(10)
```

```python
df.columns = ['lat', 'long']
```

```python
df['latlong'] = df.apply(combine_lat_long, axis=1)
df['next_lat_long'] = df.latlong.shift(-1)
df['distance_between'] = df.apply(get_distance, axis=1)
df.head(10)
```

```python
# Then you can filter
data=df.query('distance_between > .4')
data.head()
```

```python
df = pd.DataFrame(data)
df.to_csv("dist_res.csv",index=False)
```
